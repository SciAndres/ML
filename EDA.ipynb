# Modelo de prediccion 

```python
# Definir las caractersticas (variables independientes)
X = df_result[['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity']]
```

```python
# Definir la variable objetivo
y = df_result['Happiness Score']
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Inicializar el modelo de regresin lineal
modelo_regresion = LinearRegression()
```

```python
# Entrenar el modelo en los datos de entrenamiento
modelo_regresion.fit(X_train, y_train)
```

```python
# Realizar predicciones en el conjunto de prueba
y_pred = modelo_regresion.predict(X_test)
y_pred[:50]
```

# Comparacion valores reales y predicciones

```python
y_pred = modelo_regresion.predict(X_test)

# Imprimir un valor real y su prediccin correspondiente
for i in range(10):
    print(f"Valor Real: {y_test.iloc[i]}, Prediccin: {y_pred[i]}")
```

# Evaluar el rendimiento del modelo

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared (R2): {r2}')
```

# Grafica

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='red', marker='o', edgecolors='black', alpha=0.7, s=80, label='Predicciones')
plt.scatter(y_test, y_test, color='orange', marker='x', s=80, label='Valores Reales')
plt.xlabel("Valores Reales")
plt.ylabel("Predicciones")
plt.title("Grfica de Regresin")
plt.legend()
plt.show()

```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', marker='o', edgecolors='black', alpha=0.7, s=80, label='Valores Reales vs. Predicciones')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2, label='Lnea de Regresin')
plt.xlabel("Valores Reales")
plt.ylabel("Predicciones")
plt.title("Grfica de Regresin")
plt.legend()
plt.show()
```
